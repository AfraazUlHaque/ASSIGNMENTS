```python
#IDENTIFIED NUMERICAL AND CATEGORICAL ATTRIBUTES
nums_col = df.select_dtypes(include =['int64' , 'float64']).columns
cat_col = df.select_dtypes(exclude=['int64' , 'float64']).columns
print("NUMERICAL COLUMNS : " , nums_col)
print("CATEGORICAL COLUMNS : " , cat_col)
```

# **TASK 2**

```python
#MISSING VALUES
df.isnull().sum()
```

```python
#total no of missing values
df.isnull().sum().sum()
```

```python
#duplicates
df.duplicated().sum()
```

```python
#Summary statics
df.describe()
```

```python
#for categoricals
df.describe(include=['object'])
```

# **TASK 2**

```python
#REPLACING NUMERICAL DATA WITH MEDIAN
for col in nums_col:
    df[col] = df[col].fillna(df[col].median() , inplace = True )
```

```python
#REPLACING CATEGORICAL DATA WITH MODE
for col in cat_col:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
#verifying
df.isnull().sum().sum()
```

# **TASK 3**

```python
df.drop_duplicates(inplace = True)
```

```python
#SHAPE AFTER REMOVING DUPLICATES
df.shape
```
